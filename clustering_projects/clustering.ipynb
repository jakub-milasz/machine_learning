# Hierarchical Clustering in dataset Wholesale customers

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Wholesale customers data.csv')
X = dataset.iloc[:, [6, 7]].values
```

## Creating the dendrogram to choose the number of clusters

```python
import scipy.cluster.hierarchy as sch
Z = sch.linkage(X, method = 'ward')
sch.dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Euclidean distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 4)
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], color = 'blue', s = 100, label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], color = 'red', s = 100, label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], color = 'green', s = 100, label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], color = 'cyan', s = 100, label = 'Cluster 4')
plt.title('Clusters')
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')
plt.show()
```
